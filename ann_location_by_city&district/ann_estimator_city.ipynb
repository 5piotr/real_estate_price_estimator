```python
import pickle as pkl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

```python
infile = open('..//real_estate_data_frame','rb')
frame = pkl.load(infile)
infile.close()
```

```python
frame
```

```python
frame.info()
```

```python
frame = frame.drop(['district','localization_y','localization_x'],axis=1)
```

```python
frame.info()
```

## creating dummy variables

```python
dummies = pd.get_dummies(frame[['city','sewers','road','fence','water','electricity','gas']],drop_first=True)
frame = frame.drop(['city','sewers','road','fence','water','electricity','gas'],axis=1)
frame = pd.concat([frame,dummies],axis=1)
```

```python
frame.info()
```

```python
frame
```

## train test split

```python
X = frame.drop('price',axis=1).values
y = frame['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

## scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

## creating model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation ,Dropout
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()

model.add(Dense(1024,activation='relu',input_dim=X_test.shape[1]))
model.add(Dense(512,activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.summary()
```

```python
# from tensorflow.keras.callbacks import EarlyStopping
```

```python
# early_stop = EarlyStopping(monitor='val_loss',patience=2)
```

```python
start = time.time()

model.fit(x=X_train,y=y_train,
          validation_data=(X_test,y_test),
          batch_size=256,epochs=1500,
#           callbacks=[early_stop],
         )

stop = time.time()
print('\n',(stop-start)//60,'min.')
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
# plt.ylim(0,20000000000)
```

## model ewaluation

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_absolute_error(y_test,predictions)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
explained_variance_score(y_test,predictions)
```

```python
plt.figure(figsize=(7,7))
# plt.xlim(0,3000000)
# plt.ylim(0,3000000)

# Our predictions
plt.scatter(y_test,predictions,edgecolor=None,alpha=0.2)

# Perfect predictions
plt.plot(y_test,y_test,'r')
```
