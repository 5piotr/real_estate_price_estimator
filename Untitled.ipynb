```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pickle as pkl
import random
```

```python
infile = open('real_estate_data_frame_improved','rb')
dataset = pkl.load(infile)
infile.close()
```

```python
dataset.drop('price',axis=1,inplace=True)
```

```python
X = dataset.iloc[:,[0,1]].values
```

```python
X
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 200, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
```

```python
for i in range(200):
    plt.scatter(X[y_kmeans == i, 0], X[y_kmeans == i, 1], s = 10, color = (random.random(), random.random(), random.random()))

# plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 30, c = 'yellow', label = 'Centroids')
plt.title('Clusters of clients')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending score (1-100)')
plt.legend()
plt.show()
```
